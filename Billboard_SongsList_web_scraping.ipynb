# 6.	Scrape the details of top 100 songs on billboard.com. Url = https://www.billboard.com/
You have to find the following details:

A)	Song name

B)	Artist name

C)	Last week rank

D)	Peak rank

E)	Weeks on board

Note: - From the home page you have to click on the charts option then hot 100-page link through code.


```python
import selenium
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
driver = webdriver.Chrome(r"C:\Users\HP\Downloads\chromedriver_win32\chromedriver.exe")

url = "https://www.billboard.com/"
driver.get(url)
driver.maximize_window()
```

```python
chart_btn = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/nav/ul/li[3]/a").click()
```

```python
begin = time.time()

song_name=[]
artist_name=[]
last_week=[]
peak_rank=[]
week_board=[]

sng_name = driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
for sng in sng_name:
    song_name.append(sng.text)

art_name = driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
for art in art_name:
    artist_name.append(art.text)
    
lst_week = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for last in lst_week:
    last_week.append(last.text)    
    
pk_rank = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for peak in pk_rank:
    peak_rank.append(peak.text)  

wk_board = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for wboard in wk_board:
    week_board.append(wboard.text) 
    

main_list = {
        "Song Name":song_name,
        "Artist Name":artist_name,
        "Last Week Rank":last_week,
        "Peak Rank":peak_rank,
        "Weeks on Board":week_board  
        }


end = time.time()

print(f"Total runtime of the program is {end - begin}")
```

```python
pd.set_option("display.max_rows",None)
df_billboard = pd.DataFrame(main_list)
df_billboard
```
